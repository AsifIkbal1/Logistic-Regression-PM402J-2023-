```r
df <- read.csv("/kaggle/input/diabetes/DIAB.csv",
                 header = TRUE,
                 quote = "\"",
                 stringsAsFactors = TRUE,
                 strip.white = TRUE)
```

Explore dataset


```r
df
summary(df)
```

Change the diabetes column to a factor

```r
df$DIABET <- as.factor(df$DIABET)
summary(df)
```

Run logistic regression

```r
lgr <- glm(DIABET ~ AGE+BMI ,data=df , family="binomial")
```

Make a prediction

```r
patient_data <- data.frame(AGE=36,BMI=27)
patient_data
```

```r
predict(lgr, patient_data, type="response")
```

Train a model on 75% of the data, test on 25%

```r
bound <- (nrow(df)/4)*3

df <- df[sample(nrow(df)),]
dfTrain <- df[1:bound,]
dfTest <- df[(bound+1):nrow(df),]
nrow(dfTest)
```

Retrain model

```r
lgr75 <- glm(DIABET ~ . ,data=dfTrain , family="binomial")
lgr75preds <- predict(lgr75, dfTest, type="response")
```

Create a confusion matrix

```r
table(predicted = as.numeric(lgr75preds>0.5), truth = dfTest$DIABET)
```

```r
accuracy = (110 + 36)/nrow(dfTest)
accuracy
```
